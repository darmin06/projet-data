```python
import pandas as pd
import matplotlib.pyplot as plt

baseline = pd.read_csv('../results/charts/metrics_baseline.csv')
augmented = pd.read_csv('../results/charts/metrics_augmented.csv')

avg_baseline = baseline[['Recall@5', 'NDCG@5']].mean()
avg_augmented = augmented[['Recall@5', 'NDCG@5']].mean()

metrics = pd.DataFrame({
    'Metric': ['Recall@5', 'NDCG@5'],
    'Baseline': avg_baseline.values,
    'Augmented': avg_augmented.values
})

metrics.set_index('Metric').plot(kind='bar', figsize=(6,4), title='Simulated Performance Comparison')
plt.ylabel('Score')
plt.ylim(0, 0.5)
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
