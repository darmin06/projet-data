```python
import json
import pandas as pd

with open('../data/synthetic/fake_profiles.json') as f:
    profiles = json.load(f)
interactions = pd.read_csv('../data/synthetic/fake_interactions.csv')
print(profiles[:1])
print(interactions.head())
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
for _, row in interactions.iterrows():
    user = f"user_{row['userId']}"
    item = f"movie_{row['movieId']}"
    G.add_edge(user, item, label=row['interaction_type'])
pos = nx.spring_layout(G)
colors = ['lightblue' if 'user' in n else 'lightgreen' for n in G.nodes()]
nx.draw(G, pos, with_labels=True, node_color=colors)
plt.show()
```
